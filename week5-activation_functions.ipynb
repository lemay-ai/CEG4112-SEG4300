```python
# prompt: plot an image of RELU activation

import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(0, x)

x = np.linspace(-5, 5, 100)
y = relu(x)

plt.plot(x, y)
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
# prompt: stack these three figures into one figure for inclusion in a slide

import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(0, x)

def tanh(x):
  return np.tanh(x)

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

x = np.linspace(-5, 5, 100)

fig, axes = plt.subplots(3, 1, figsize=(8, 10))

# ReLU
y_relu = relu(x)
axes[0].plot(x, y_relu)
axes[0].set_title('ReLU Activation Function')
axes[0].set_xlabel('Input')
axes[0].set_ylabel('Output')
axes[0].grid(True)

# Tanh
y_tanh = tanh(x)
axes[1].plot(x, y_tanh)
axes[1].set_title('Tanh Activation Function')
axes[1].set_xlabel('Input')
axes[1].set_ylabel('Output')
axes[1].grid(True)

# Sigmoid
y_sigmoid = sigmoid(x)
axes[2].plot(x, y_sigmoid)
axes[2].set_title('Sigmoid Activation Function')
axes[2].set_xlabel('Input')
axes[2].set_ylabel('Output')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment="Neuron Diagram")

# Add input nodes
inputs = ['in1', 'in2', 'in3', '...', 'inn']  # List of input names
weights = ['w1', 'w2', 'w3', '...', 'wn']    # Corresponding weights

for i in range(len(inputs)):
    dot.node(inputs[i], f"{inputs[i]}")  # Add input nodes

# Add bias node
dot.node("bias", "bias", shape="circle", style="filled", fillcolor="lightgray")

# Add summation node
dot.node("sum", "", shape="circle", style="bold", fontsize="16")

# Add activation node
dot.node("activation", "Activation", shape="rect", style="rounded,filled", fillcolor="lightblue")

# Add output node
dot.node("output", "Output", shape="rect", style="rounded,filled", fillcolor="lightgreen")

# Connect input nodes to the summation node via weights
for i, input_name in enumerate(inputs):
    weight_label = weights[i] if i < len(weights) else "..."
    dot.edge(input_name, "sum", label=weight_label)

# Connect bias to the summation node
dot.edge("bias", "sum", label="+", style="dashed")

# Connect summation node to activation function
dot.edge("sum", "activation")

# Connect activation function to output
dot.edge("activation", "output")

# Render the graph
dot.render("neuron_diagram", format="png", cleanup=True)  # Generates neuron_diagram.png
print(dot.source)
```
