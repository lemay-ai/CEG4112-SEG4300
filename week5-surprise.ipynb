```python
# Step 1: Install Surprise 
# Note: There is NumPy/scikit-surprise compatibility issue for 'numpy>2', easiest solution will be to downgrade to numpy<2
# So first:
# !pip uninstall -y numpy
# !pip install numpy==1.26.4
# !pip install scikit-surprise
# Next, restart your kernel

from surprise import SVD
from surprise import Dataset
from surprise.model_selection import train_test_split

# Step 2: Load the built-in MovieLens 100k dataset
data = Dataset.load_builtin('ml-100k')

# Step 3: Split into train/test sets
trainset, testset = train_test_split(data, test_size=0.2)

# Step 4: Train a model using SVD
algo = SVD()
algo.fit(trainset)

# Step 5: Make a prediction
# Example: Predict rating for user '196' on item '302'
prediction = algo.predict(uid='196', iid='302')
print(prediction)
```
